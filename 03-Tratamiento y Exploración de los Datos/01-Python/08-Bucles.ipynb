````python
## Programa que se detenga hasta que el usuario de -1 y d el nmero de calificaciones aprobadas y reprobadas. Calificacin mnima de 7
aprobadas = 0
reprobadas = 0

while True:
    calif = float(input("Inserta una calificacin (-1 para terminar): "))

    if calif == -1:
        break
    elif calif >= 7 and calif <= 10:
        aprobadas += 1
    elif calif >= 0 and calif < 7:
        reprobadas += 1
    else:
        print("Calificacin invlida. Debe estar entre 0 y 10.")

print(f"\nPrograma terminado.")
print(f"Total de calificaciones aprobadas: {aprobadas}")
print(f"Total de calificaciones reprobadas: {reprobadas}")
````

### Bucle con FOR

Otro tipo de bucle disponible en `Python` proviene de la observacin de que a veces es ma importante contar los `giros` o `vueltas` del bucle que verificar las condiciones
```python
for i in range(100):
    # do_something()
    pass
```

````python
for i in range(10):
    print (f'el valor de i es {i}')
````

````python
for i in range(10):
    print (f'el valor de i es {i + 1}')
````

````python
lista  = [2,3,4,9,6,7,10,[2,3,5,6]]

for li in lista:
    print (li)
````

Nota:
El bucle se ha ejecutado 10 veces (es el argumento de la funcin range()). El valor de la ltima variable de control es 9 (no en 10, ya que comienza en 0, no desde 1)

````python
for i in range (2, 8):
    print (f'El valor de i es {i}')
````

````python
i = 2
while i < 8:
    print(f'El valor de i es {i}')
    i += 1
````

La funcin range tambin puede aceptar 3 argumentos ([rango inicial], [rango final], [de cunto en cunto])

````python
for i in range (2, 8, 3):
    print (f'El valor de i es: {i}')
````

````python
i = 2
while i < 8:
    print(f'El valor de i es: {i}')
    i += 3
````

## Las sentiencias break y continue

`break` - Sale del bucle inmediatamente e incodicionalmente termina la operacin del bucle. El programa comienza a ejecutar la instruccin ms cercana despus del cuerpo del bucle

````python
print('La instruccin break')
for i in range (1, 6):
    if i == 3:
        break
    print (f'Dentro del bucle {i}')
print ('Fuera del bucle')
````

````python
print('La instruccin break')
for i in range (1, 6):
    print (f'Dentro del bucle {i}')
    if i == 3:
        break
print ('Fuera del bucle')
````

`continue` - Se comporta como si el programa hubiera llegado repentinamente al final de cuerpo. El siguiente turno se inicia y la expresin de condicin se prueba de inmediato.

````python
print ('La instruccin continue')
for i in range(1 , 6):
    if i == 3:
        continue
    print (f'Dentro del bucle: {i}')
print ('Fuera del bucle')
````
