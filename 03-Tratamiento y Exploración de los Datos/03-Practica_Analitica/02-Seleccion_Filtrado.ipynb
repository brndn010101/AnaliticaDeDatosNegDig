# Filtrado con Operadores Lógicos
- `&` AND ->Devuelve filas que cumplen todas las condiciones
- `|` OR
- `~` NOT -> Devuelve filas que no cumplen con la condición

### Filtrado con Operadores Lógicos
- `&` (condicion1) & (condicion2)
- `|` (condicion1) | (condicion2)
- `~` (condicion)

```python
#Trabajos que no sean remotos
df['onsite_remote'] == 'remote'
```

```python
df[df['onsite_remote'] == 'remote']
```

Muestra todos los 'onsite_remote' que no son remotos

```python
df[~(df['onsite_remote'] == 'remote')]
```

```python
df[~((df['company'] == 'Amazon') | (df['onsite_remote'] == 'remote'))]
```

```python
df[~((df['company'] == 'Amazon') & (df['onsite_remote'] == 'remote'))]
```
